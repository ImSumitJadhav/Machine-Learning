```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/AlmaBetter/Project 2/SeoulBikeData.csv",encoding="latin1",usecols=["Seasons","Holiday","Functioning Day","Hour"])
df.head()
```

# **Feature Encoding**

# **1.One hot encoding**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
# transforming the column after fitting
df1=df.copy()
ohe_df = ohe.fit_transform(df1[['Functioning Day']]).toarray()
ohe_df=pd.DataFrame(ohe_df)
df1=pd.concat([df1,ohe_df],axis=1)
df1.head()
```

# **2.get_dummies method**

```python
import pandas as pd
df2=df.copy()
df2 = pd.get_dummies(df2, prefix=['yes',"yes"], columns=['Seasons','Holiday'])
df2.head(10)
```

# **3.Label Encoding**

```python
df3=df.copy()

# labelEncoder present in scikitlearn library
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
df3['Holiday'] = le.fit_transform(df3['Holiday'])

df3.tail(10)
```

# **4.OrdinalEncoder**

```python
from sklearn.preprocessing import OrdinalEncoder
ord1 = OrdinalEncoder()

df4=df.copy()
# fitting encoder
oe_df=ord1.fit_transform(df4[['Seasons']])
oe_df=pd.DataFrame(oe_df,columns=["Season"])
df4=pd.concat([df4,oe_df],axis=1)
df4.head()
```

# **5.Manually assigning ranking by using a dictionary**

```python
df5=df.copy()

# creating a dictionary
sea_dict = {'Spring': 1, 'Autumn': 2, 'Summer': 3,'Winter':4}
# mapping values in column from dictionary
df5['new_season'] = df5.Seasons.map(sea_dict)
df5 = df5.drop(['Seasons'], axis=1)
df5.head()
```

# **6.Binary Encoder**

```python
df6=df.copy()

from category_encoders import BinaryEncoder
encoder = BinaryEncoder(cols =['Holiday'])
# transforming the column after fitting
newdata = encoder.fit_transform(df6['Holiday'])
# concatenating dataframe
df6 = pd.concat([df6, newdata], axis = 1)
# dropping old column
df6 = df6.drop(['Holiday'], axis = 1)
df6.head(10)
```

# **7.Hash Encoding**

```python
df7=df.copy()

from sklearn.feature_extraction import FeatureHasher
# n_features contains the number of bits you want in your hash value.
h = FeatureHasher(n_features = 4, input_type ='string')
# transforming the column after fitting
hashed_Feature = h.fit_transform(df7['Holiday']).toarray()

df7 = pd.concat([df7, pd.DataFrame(hashed_Feature)], axis = 1)
df7.head()
```

# **8.Multilabel binerizer**

```python
df8=df.copy()

from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
mlb_array=mlb.fit_transform(df8["Seasons"])
mlb_df=pd.DataFrame(mlb_array)
df8=pd.concat([df8,mlb_df],axis=1)
df8.head(10)
```

# **Feature Selection Techiniques**

# **1.Pearson feature selection (f_regression)**

```python
# pearson's correlation feature selection for numeric input and numeric output
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

# define feature selection
fs = SelectKBest(score_func=f_regression, k=5)
# apply feature selection
X1_selected = fs.fit_transform(x, y)
print(X1_selected.shape)

X1_selected=pd.DataFrame(X1_selected)
X1_selected.head(10)
```

# **2.ANOVA feature selection (f_classif)**

```python
# ANOVA feature selection for numeric input and categorical output
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# define feature selection
fs = SelectKBest(score_func=f_classif, k=6)
# apply feature selection
X2_selected = fs.fit_transform(x, y)
print(X2_selected.shape)

X2_selected=pd.DataFrame(X2_selected)
X2_selected.head(10)
```

# **3.Feature Importance with tree (ExtraTreesRegressor or ExtraTreesClassifier)**

```python
# Feature Importance with Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier

# feature extraction
model = ExtraTreesClassifier(n_estimators=10)
model.fit(x2, y2)
imp=model.feature_importances_
imp=pd.DataFrame(imp)
imp.head(20)
```

# **4.Principal Component Analysis**

```python
# Feature Extraction with PCA
from sklearn.decomposition import PCA

# feature extraction
pca = PCA(n_components=3)
fit = pca.fit(x2)
# summarize components

a1=fit.components_
a1=pd.DataFrame(a1)
a1.head(20)
```

# **5.chi2 selection**

```python
from  sklearn.feature_selection import chi2

# feature extraction
model2 = chi2(x2,y2)

a2=pd.DataFrame(model2)
a2.head(20)
```

# **6.Mutual Information (mutual_info_regression or mutual_info_classif)**

```python
from sklearn.feature_selection import mutual_info_regression

imp2=pd.DataFrame(mutual_info_regression(x,y))
imp2
```
